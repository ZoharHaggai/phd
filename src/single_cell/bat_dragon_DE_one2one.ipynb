# Single Cell analysis bat dragon

```python
from itertools import product
from pathlib import Path

import pandas as pd
import numpy as np
```

```python
# Trying things on the bat_dragon_DE file from the drive

# animal = 'bat'
animal = 'komodo'

# sample_type = 'lps'
sample_type = 'pic'


base_filepath = f'../../data/single_cell/animals/{animal}/{sample_type}/{animal}_{sample_type}.csv'
print(base_filepath)
```

```python
df_komodo_pic = pd.read_csv(base_filepath, index_col=0)
df_komodo_pic.sort_index()

df_komodo_pic
```

#### Filtering out rows that are not 1 to 1 ortholgs (has 'no' in the 1to1_bat_dragon column)

```python
# Filter out the row where 'no' appears

filtered_df_komodo_pic = df_komodo_pic[df_komodo_pic['1to1_bat_dragon'] != 'no']
filtered_df_komodo_pic
```

```python
pvalue_adj_sig = filtered_df_komodo_pic[filtered_df_komodo_pic[f'{sample_type}_pvals_adj'] < 0.05]
komodo_pic_logfc_up = pvalue_adj_sig[pvalue_adj_sig[f'{sample_type}_logfoldchanges'] >= 0]
komodo_pic_logfc_up
```

```python
komodo_pic_logfc_up.to_csv(f'../../data/single_cell/animals/{animal}/{sample_type}/{animal}_{sample_type}_logfc_up_DE.csv')
komodo_pic_logfc_up
```
