```python
from itertools import product
from pathlib import Path

import pandas as pd
import numpy as np
import os
```

```python
# Trying things on the bat_dragon_DE file from the drive

animal_b = 'bat'
animal_k = 'komodo'

sample_type = 'lps'
# sample_type = 'pic'

# bat DF

df_bat_lps = pd.read_csv(f'../../data/single_cell/animals/{animal_b}/{sample_type}/{animal_b}_{sample_type}.csv', index_col=0)
df_bat_lps = df_bat_lps.sort_index()

df_bat_lps
```

```python
# komodo DF

df_komodo_lps = pd.read_csv(f'../../data/single_cell/animals/{animal_k}/{sample_type}/{animal_k}_{sample_type}.csv', index_col=0)
df_komodo_lps = df_komodo_lps.sort_index()

df_komodo_lps
```

```python
df_combined = df_bat_lps.merge(df_komodo_lps, left_index=True, right_index=True, how='outer')

df_combined
```

```python
# filtering by taking out the rows that says 'no' in the column 1to1_bat_dragon_x (x presents the bat's df)

filtered_df_combined_x = df_combined[df_combined['1to1_bat_dragon_x'] != 'no']

filtered_df_combined_x
```

```python
pvalue_adj_sig = filtered_df_combined_x[filtered_df_combined_x[f'{sample_type}_pvals_adj_x'] < 0.05]
combined_lps_logfc_up = pvalue_adj_sig[pvalue_adj_sig[f'{sample_type}_logfoldchanges_x'] >= 0]
combined_lps_logfc_up.sort_values(by=['lps_logfoldchanges_x'], ascending=False, inplace=True)

combined_lps_logfc_up
```

```python
combined_lps_logfc_up.to_csv(f'../../data/single_cell/animals/{animal_b}/{sample_type}/{animal_b}_{sample_type}_logfc_up_combined.csv')

combined_lps_logfc_up
```
